```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql import functions as F
import pandas as pd
from pandas import read_csv
from datetime import datetime
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
```

```python
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
```

```python
Bitcoin = spark.read.format('csv').option("escape","\"").option('header',True).load("Bitcoin.csv")
DowJones = spark.read.format('csv').option("escape","\"").option('header',True).load("DowJones.csv")
SP500 = spark.read.format('csv').option("escape","\"").option('header',True).load("SP500.csv")
Nasdaq = spark.read.format('csv').option("escape","\"").option('header',True).load("Nasdaq.csv")
```

```python
Bitcoin=Bitcoin.toPandas()
DowJones=DowJones.toPandas()
SP500=SP500.toPandas()
Nasdaq=Nasdaq.toPandas()
```

```python
Bitcoin['Date']= pd.to_datetime(Bitcoin['Date']) 
Bitcoin['Open'] = Bitcoin['Open'].astype(float)
Bitcoin['High'] = Bitcoin['High'].astype(float)
Bitcoin['Low'] = Bitcoin['Low'].astype(float)
Bitcoin['Close'] = Bitcoin['Close'].astype(float)
Bitcoin['Rate'] = Bitcoin['Rate'].astype(float)
Bitcoin['Rate'] = Bitcoin['Rate']*-1
```

```python
DowJones['Date']= pd.to_datetime(DowJones['Date']) 
DowJones['Open'] = DowJones['Open'].astype(float)
DowJones['High'] = DowJones['High'].astype(float)
DowJones['Low'] = DowJones['Low'].astype(float)
DowJones['Close'] = DowJones['Close'].astype(float)
DowJones['Rate'] = DowJones['Rate'].astype(float)
DowJones['Rate'] = DowJones['Rate']*-1
```

```python
SP500['Date']= pd.to_datetime(SP500['Date']) 
SP500['Open'] = SP500['Open'].astype(float)
SP500['High'] = SP500['High'].astype(float)
SP500['Low'] = SP500['Low'].astype(float)
SP500['Close'] = SP500['Close'].astype(float)
SP500['Rate'] = SP500['Rate'].astype(float)
SP500['Rate'] = SP500['Rate']*-1
```

```python
Nasdaq['Date']= pd.to_datetime(Nasdaq['Date']) 
Nasdaq['Open'] = Nasdaq['Open'].astype(float)
Nasdaq['High'] = Nasdaq['High'].astype(float)
Nasdaq['Low'] = Nasdaq['Low'].astype(float)
Nasdaq['Close'] = Nasdaq['Close'].astype(float)
Nasdaq['Rate'] = Nasdaq['Rate'].astype(float)
Nasdaq['Rate'] = Nasdaq['Rate']*-1
```
